# Análise do experimento

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import math as m
import statsmodels.api as sm
```

## Importando os dados e separando as variáveis

```python
df_platina = pd.read_excel('platina.xlsx') # Dados obtidos para o fio de platina (condutor)
df_GaAs = pd.read_excel('GaAs.xlsx') # Dados obtidos para o arsenieto de gálio (semicondutor)
```

```python
temperatura_platina = df_platina['Temperatura_kelvin']
resistencia_platina = df_platina['Resistência_ohm']

temperatura_GaAs = df_GaAs['Temperatura_kelvin']
resistencia_GaAs = df_GaAs['Resistência_ohm']
```

## Realizando uma regressão linear sobre os dados coletados para o fio de platina (condutor)

A equação 8 prevê que neste caso, teremos uma relação linear.

```python
x1 = sm.add_constant(temperatura_platina)
y1 = resistencia_platina
results1 = sm.OLS(y1,x1).fit()
```

```python
results1.summary()
```

##  Resistência em função da temperatura, condutor e semicondutor

```python
sns.set(rc={"figure.figsize": (15, 7.5)})
plt.subplot(1,2,1)    
plt.title('Fio de platina',fontsize=20)
plt.scatter(temperatura_platina, resistencia_platina)
plt.plot(temperatura_platina, results1.predict(x1), c='red', label = 'Regressão linear')
plt.legend(loc=[0.01,0.92],fontsize=16)
plt.xlabel('Temperatura (K)', fontsize=16)
plt.ylabel('Resistência (ohm)',fontsize=16)
        
plt.subplot(1,2,2)     
plt.title('Arsenieto de gálio (GaAs)',fontsize=20)
plt.scatter(temperatura_GaAs, resistencia_GaAs)
plt.xlabel('Temperatura (K)', fontsize=16)
plt.ylabel('Resistência (ohm)',fontsize=16)

plt.show()
```



O crescimento linear da resistência com a temperatura, previsto pela equação 8 para um condutor, é validado no gráfico obtido para o fio de platina, gerado na célula 6. Assim como a queda exponencial prevista pela equação 6, para a resistência de um semicondutor com a temperatura, é também validado para o gráfico obtido para o arsenieto de gálio.

O coeficiente de inclinação $a$, obtido em uma regressão linear realizada sobre os dados obtidos para o fio de platina, foi $a=(0.3688 \pm 0.001) ohm/K$. A equação 7 que descreve $\lambda$, fator que gerou a proporcionalidade da resistência com a temperatura, mostra que $\lambda$ é proporcional a massa dos íons $M$, o que gera uma proporcionalidade entre a resistência $R$, e o inverso da massa dos íons $1/M$. Por este fato, se o experimento fosse realizado com condutores de massa atômica menores que a platina [(195.084 u)](https://pt.wikipedia.org/wiki/Platina), como o alumínio [(26.9815386 u)](https://pt.wikipedia.org/wiki/Alum%C3%ADnio) e o cobre [(63.546 u)](https://pt.wikipedia.org/wiki/Cobre) por exemplo, esperaríamos que o coeficiente de inclinação $a$ assumisse valores maiores.

Para o caso do arsenieto de gálio, a equação 6 prevê que se plotarmos o logaritmo da resistência, em função do inverso da temperatura, teremos uma relação linear, e o coeficiente de inclinação obtido a partir de uma regressão linear sobre os dados, será uma medida direta da energia de ativação $E_A$, apenas dividido pela constante de Boltzmann $K_B$. O resultado deste procedimento pode ser visualizado no gráfico gerado na célula 11. A energia de ativação $E_A$ medida, foi gerada na célula 12.



## Logaritmo da resistência em função do inverso da temperatura, semicondutor

```python
def inverso_T(t):
    return 1/t
```

```python
inverso_temperatura_GaAs = temperatura_GaAs.apply(inverso_T)
ln_resistencia_GaAs = resistencia_GaAs.apply(m.log)
```

```python
x = sm.add_constant(inverso_temperatura_GaAs)
y = ln_resistencia_GaAs
results = sm.OLS(y,x).fit()
```

```python
results.summary()
```

```python
sns.set(rc={"figure.figsize": (7.5, 7.5)})
plt.subplot(1,1,1) 
plt.title('Arsenieto de gálio (GaAs)',fontsize=20)
plt.scatter(inverso_temperatura_GaAs, ln_resistencia_GaAs)
plt.plot(inverso_temperatura_GaAs, results.predict(x), c='red', label = 'Regressão linear')
plt.legend(loc=[1.01,0.93],fontsize=16)
plt.xlabel('Inverso da temperatura ($K^-1$)',fontsize = 16)
plt.ylabel('logaritimo da resistência (log(ohm))', fontsize = 16)

plt.show()
```
